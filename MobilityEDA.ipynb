Exploring BNE to MEL first

```python
#most popular time of year for brissy --> melbs travellers 
year_map = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec",
}
BNE_to_MEL_df["month_name"] = BNE_to_MEL_df["month"].map(year_map)
sns.countplot(x="month_name", data=BNE_to_MEL_df, order=year_map.values())
plt.xlabel("Month")
plt.ylabel("Number of travellers")
```

```python
#most visited spots
BNE_to_MEL_df["visited_lga"].value_counts().plot.barh()
plt.title("Most Popular Spots to Visit in MEL Over 3 years")
plt.xlabel("frequency")
```

```python
BNE_to_MEL_df[BNE_to_MEL_df["visited_lga"] == "Melbourne"].groupby("year")["visitors"].sum().plot()
plt.title("Popularity of years in Melbourne (most popular)")
plt.xticks(np.arange(2020, 2023))
plt.ylabel("no. visitors")
```

```python
#Check out most popular destination first
melbs = BNE_to_MEL_df[BNE_to_MEL_df["visited_lga"] == "Melbourne"]
melbs
```

```python
sns.pairplot(melbs)
```

```python
melbs.describe()
```

```python
#average income for people travelling to melbs 
melbs["median_wkly_family_income"].mean()
```

```python
melbs_corr = melbs.corr(method ="pearson")
melbs_corr
```

```python
sns.heatmap(melbs_corr)
```

```python
melbs_wk_income = pd.DataFrame(melbs["median_wkly_family_income"]).copy()
melbs_visitors = pd.DataFrame(melbs["visitors"]).copy()
melbs_month = melbs["month"].copy()

concat_= pd.concat([melbs_wk_income, melbs_visitors, melbs_month], axis = 1)
concat_
```

```python
concat_.describe()
```

```python
#box plot for melbs income 
melbs_income_box = pd.DataFrame(concat_["median_wkly_family_income"])
melbs_income_box.boxplot()
```

```python
sns.lineplot(data=concat_, x="month", y="median_wkly_family_income", ci=None)
plt.xticks(np.arange(1, 13, 1))

plt.title("Median Weekly household income of travellers to Melbourne")

```

So we can understand the median income of families travelling to melbourne for each month and what times families with the highest income travel at. 

```python
melbs_travel_pop = pd.DataFrame(melbs.iloc[:, 11:19]).copy()
melbs_travel_pop
```

```python
sns.pairplot(melbs_travel_pop)
```

```python
melbs_travel_pop.corr(method ="pearson")
```

```python
pos_demographic_ladies = melbs_travel_pop['age_25_34_yr'].corr(melbs_travel_pop['region_female_pop'])
pos_demographic_ladies
```

```python
pos_demographic_men = melbs_travel_pop['age_25_34_yr'].corr(melbs_travel_pop['region_male_pop'])
pos_demographic_men 
```

So we know that the highest positively correlated pair travelling to Melbourne is men and ladies aged 25-34. This makes alot of since considering current culture and financial standpoints. This is important to know the demographic of travellers coming to area for business - target young families 

```python
#check out least popular melb destination 
maroondah_df = pd.DataFrame(BNE_to_MEL_df[BNE_to_MEL_df["visited_lga"] == "Maroondah"]).copy()
maroondah_df
```

```python
maroondah_travel_pop = pd.DataFrame(maroondah_df.iloc[:, 11:19]).copy()
maroondah_travel_pop 
```

```python
maroondah_travel_pop .corr(method ="pearson")
```

So same thing, most correlated travels are men and women 25-34

```python
fig1 = plt.figure(figsize=(12, 12))
ax = fig1.add_subplot(1, 1, 1)
plt.xticks(np.arange(1, 13, 1))
BNE_to_MEL_df[BNE_to_MEL_df["visited_lga"] == "Melbourne"].groupby("month")["visitors"].sum().plot()
BNE_to_MEL_df[BNE_to_MEL_df["visited_lga"] == "Maroondah"].groupby("month")["visitors"].sum().plot()
plt.legend(['Melbourne travellers', 'Maroondah travellers'])
plt.title("Frequency Travellers to Melbourne (most popular) and Maroondah (least popular) destinations")
plt.ylabel("No. Visitors")
```

```python
fig1 = plt.figure(figsize=(12, 12))
ax = fig1.add_subplot(1, 1, 1)
sns.lineplot(data=concat_, x="month", y="median_wkly_family_income", ci=None)
sns.lineplot(data=maroondah_df, x="month", y="median_wkly_family_income", ci=None)
ax.legend(['Melbourne travellers', 'Maroondah travellers'])
plt.xticks(np.arange(1, 13, 1))
plt.title("Median Family Income of Travellers to Melbourne (most popular) and Maroondah (least popular) destinations")
```
