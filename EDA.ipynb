```python
"""
Using a big pairplot or draftsman plot. It's a method for looking at the 
interrelations between variables in multivariate data.
The plot consists of a simple array of two-variable scatter diagrams.

"""
sns.pairplot(df)
```

```python
#plot histograms
df.hist(figsize=(12, 12), bins=15)
```

```python
#minimize dataset and take key features with significant relationships

mini_df = df[["Unnamed: 0", 'visitors' ,'device_weight', 'estimate']].copy()
mini_df.describe()
```

```python
#reduced pairplot for key relationships
sns.pairplot(mini_df)
```

```python
#correlation between key features

datacorr = mini_df.corr(method ="pearson")
datacorr
```

```python
#visualise relationships 

sns.heatmap(datacorr)
```

```python
#highest positively correlated pair
mini_df['device_weight'].corr(mini_df['estimate'])
```

```python
#highest negatively correlated pair
mini_df['device_weight'].corr(mini_df['visitors'])
```

```python
mini_df
```

```python

device_weight = mini_df["device_weight"]
estimate = mini_df["estimate"]
plt.boxplot ([device_weight])
plt.xlabel("device weight")
plt.ylabel("units of weight")
plt.title("Box plot of Device Weight")
```

```python
device_weight.describe()
```

```python
plt.boxplot ([estimate])
plt.xlabel("estimate")
plt.ylabel("estimation units")
plt.title("Box plot of Estimations Made")
```

```python
estimate.describe()
```

```python
#Linear Regression model for most correlated pair
x = mini_df['estimate']
y = mini_df['device_weight']
plt.scatter(x, y)
plt.title("Linear Regression Model for Device Weight and Estimate")
plt.xlabel('device_weight')
plt.ylabel('estimate')

r = np.polyfit(x, y, 1)
p = np.poly1d(r)
plt.plot(x, p(x), "r--")
plt.show()
```
